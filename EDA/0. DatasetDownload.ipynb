```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
from os import makedirs
import os
```

```python
kaggle_dict = {'it23chicagocrime2026':'ChicagoCrimes(20152025).csv',
               'it23chicagoarea':'ChicagoCommunityArea.csv'}

makedirs("../ProjectData/",exist_ok=True)


folder_path = "../ProjectData/"
account = 'rkyz801/'

for k,v in kaggle_dict.items():

    kaggle_dir = account + k
    local_file_path = folder_path + v

    if os.path.exists(local_file_path) and os.path.exists(local_file_path.replace(".csv",".parquet")):
        pass
        print(f"{v} and {v.replace(".csv",'.parquet')}  exists in folder ProjectData")
    else:
        df = kagglehub.load_dataset(
        KaggleDatasetAdapter.PANDAS,
        account + k,
        v,
        )
        df.to_csv(local_file_path)
        df.to_parquet(local_file_path.replace(".csv",".parquet"))
        print(f"{v} and {v.replace(".csv",'.parquet')}  has been downloaded and saved into folder ProjectData")
```
