```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
from os import makedirs
import os

kaggle_dict = {'it23chicagocrime2026':'ChicagoCrimes(20152025).csv',
               'it23chicagoarea':'ChicagoCommunityArea.csv'}

makedirs("../ProjectData/",exist_ok=True)


folder_path = "../ProjectData/"
account = 'rkyz801/'

for k,v in kaggle_dict.items():

    kaggle_dir = account + k
    local_file_path = folder_path + v

    if os.path.exists(local_file_path):
        pass
        print(v + " exists in -> " + local_file_path)
    else:
        df = kagglehub.load_dataset(
        KaggleDatasetAdapter.PANDAS,
        account + k,
        v,
        )
        df.to_csv(local_file_path)
        print(v + " has been downloaded and saved -> " + local_file_path)
```
